## 1. What are the airlines in the dataset, accompanied by their frequencies?

```python
#to check the no of airlines
df['airline'].nunique()
```

```python
#to get their names
df['airline'].unique()
```

```python
# showing all airlines with their frequencies
df['airline'].value_counts()
```

```python
# showing the above result in bar chart
df['airline'].value_counts(ascending=True).plot.barh(color=['lightgreen','lightblue'])

plt.title("Airline Frequency")
plt.xlabel('No of Flights ')
plt.ylabel(' Airlines')
plt.show()
```

## 2. Show Bar Graphs representing the Departure Time & Arrival Time.

```python
df.head()
```

```python
#showing the dep time
df['departure_time'].value_counts()
```

```python
#showing Aarival time
df['arrival_time'].value_counts()
```

```python
#showing the dep time and arrival time for the flights with their counts
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(df['departure_time'].value_counts().index,df['departure_time'].value_counts().values,color=['lightblue','lightgreen'])
plt.title('Departure_Time')
plt.xlabel('D.Time')
plt.ylabel('D.Freq')

plt.subplot(1,2,2)
plt.bar(df['arrival_time'].value_counts().index,df['arrival_time'].value_counts().values,color=['lightgreen','yellow'])
plt.title('arrival_time')
plt.xlabel('A.Time')
plt.ylabel('A.Freq')

plt.show()
```

## Does price varies with airlines ?

```python
df.head()
```

```python
#group by the airlines and check their mean price
df.groupby('airline')['price'].mean()
```

```python
# showing categorical plot showing the mean ticket price for each airlines
sns.catplot(x='airline',y='price',kind='bar',palette='coolwarm',data=df,hue='class')
plt.show()
```
